#### ii) Centralize the network
    Remove the friends of friends from the network

```python
def centralize(friend_network:nx.Graph, my_id:str = 'bilalansari.000') -> nx.Graph:
    my_friends = list(friend_network.neighbors(my_id))
    for friend in my_friends:
        matualfriends = np.intersect1d(my_friends, list(friend_network.neighbors(friend)))
        non_matualfriends = np.setdiff1d(list(friend_network.neighbors(friend)), matualfriends)
        friend_network.remove_nodes_from(non_matualfriends)
    return friend_network
```

```python
friend_network = centralize(friend_network) 
friend_network.remove_node('bilalansari.000')
friend_network = friend_network.to_undirected()
```

#### iii) Remove the friends having matual friends less than 4

```python
def removeNodes(friend_network:nx.Graph, deg: int = 4) -> nx.Graph:
    remove = [node for node,degree in dict(friend_network.degree()).items() if degree < deg]
    friend_network.remove_nodes_from(remove)
    return friend_network
```

```python
friend_network = removeNodes(friend_network)
```

```python
#Descriptive statistics
def get_descriptive_stat(graph:nx.Graph)->dict:
    stat = {}
    degree_sequence = list(dict(graph.degree()).values())
    stat['Number of nodes'] = len(graph.nodes())
    stat['Number of edges'] = len(graph.edges())
    stat['Maximum degree'] = np.max(degree_sequence)
    stat['Minimum degree'] = np.min(degree_sequence)
    stat['Std degree'] = np.std(degree_sequence)
    stat['Average degree'] = np.mean(degree_sequence)
    stat['Median degree'] = np.median(degree_sequence)
    return stat
```

```python
data = {}
for name,graph in [('My SN',friend_network),('Erdos-Rnyi',graph_erdos),('Barabasi-Albert',graph_barabasi)]:
    stat = get_descriptive_stat(graph)
    index = list(dict(stat).keys())
    data[name] = list(dict(stat).values())
pd.DataFrame(data, index = index).T
```

```python
import re
def find_friend_from_url(url):
        """Fetch the friend's facebook id from url"""
        if re.search('com\/profile.php\?id=\d+\&', url) is not None:
            m = re.search('com\/profile.php\?id=(\d+)\&', url)
            friend_id = m.group(1)
        else:
            m = re.search('com\/(.*)\?', url)
            friend_id = m.group(1)
        return friend_id
```

```python
import pickle
from tqdm.notebook import tqdm

def get_more_info(path='data/more_info.pickle'):
    with open('data/friends_url.pickle', 'rb') as f:
        friends_url = pickle.load(f)
        more_info = {}
        
        if os.path.isfile(path):
            with open(path, 'rb') as f:
                more_info = pickle.load(f)
        
        count = 1
        for url in tqdm(friends_url, desc="Fatching more information......"):
            if count % 100 == 0:
                print ("Too many queries, pause for a while...")
                time.sleep(1800)
            friend_info = {}
            friend_id = find_friend_from_url(url)
            friend_id = "bilalansari2341234.000"
            if friend_id in more_info.keys():
                continue
            friend_aboul_url = 'https://www.facebook.com/{}/about'.format(friend_id)
            driver.get(friend_aboul_url)
            friend_info['name'] = driver.title[4:]
            driver.execute_script("window.scrollTo(0, 300);")
            time.sleep(3)
            for elem_name in ["Followers", "Following"]:
                try:
                    element =  driver.find_element_by_name(elem_name)
                    friend_info[elem_name.lower()] = element.text[9:]
                except:
                    friend_info[elem_name.lower()] = "None"
            
            for link_text, elem_info in links_info.items():
                try:
                    link = driver.find_element_by_partial_link_text(link_text)
                    driver.get(link.get_attribute('href'))
                except:
                    pass
                for elem_id, elem_name in elem_info:
                    try:
                        if elem_id == "pagelet_eduwork" and friend_id == "bilalansari.000":
                            elem_id = "pagelet_edit_eduwork"
                        element = driver.find_element_by_id(elem_id) 
                        friend_info[elem_name] = element.text
                    except Exception as e:
                        friend_info[elem_name] = "None"
            try:
                link = driver.find_element_by_partial_link_text("Life events")
                driver.get(link.get_attribute('href'))
                element = driver.find_element_by_class_name('_4qm1')
                friend_info['life_events'] = element.text
            except Exception as e:
                friend_info['life_events'] = "None"
            
            more_info[friend_id] = friend_info
            with open(path, 'wb') as f:
                pickle.dump(more_info, f)
            time.sleep(5)
            count += 1
        return more_info
```

```python
import networkx as nx
import pickle

def convert(to = 'gml'):
    with open("data/friends_network.pickle", 'rb') as f:
        friends_network = pickle.load(f)
        nodes = friends_network.keys()
        edges = [(friend, friend_of_friend) 
                  for friend, friends_of_friend in friends_network.items() for friend_of_friend in friends_of_friend]
        G = nx.Graph()
        G.add_nodes_from(nodes)
        G.add_edges_from(edges)
        
        write = {'gml': lambda: nx.write_gml(G, 'data/friends_network.gml'), 
                 'graphml': lambda: nx.write_graphml(G, 'data/friends_network.graphml')}
        read = {'gml': lambda: nx.read_gml('data/friends_network.gml'), 
                 'graphml': lambda: nx.read_graphml(G, 'data/friends_network.graphml')}

        write[to]()
        return read[to]()
```

```python
class MyHTMLParser(HTMLParser):
    urls = []

    def error(self, message):
        print("Error, while parsing the HTML source code..")

    def handle_starttag(self, tag, attrs):
        # Only parse the 'anchor' tag.
        if tag == "a":
            # Check the list of defined attributes.
            for name, value in attrs:
                # If href is defined, print it.
                if name == "href":
                    if re.search('\?href|&href|hc_loca|\?fref', value) is not None:
                        if re.search('.com/pages', value) is None:
                            self.urls.append(value)
```

```python
class FBFriendData:
    def __init__(self, user:str, pwd:str = None, manual_login:bool = False):
        chrome_options = webdriver.ChromeOptions()
        prefs = {"profile.default_content_setting_values.notifications": 2}
        chrome_options.add_experimental_option("prefs", prefs)
        self.driver = webdriver.Chrome(options=chrome_options)
        self.SCROLL_PAUSE_TIME = 2
        
        self.user = user
        self.pwd = pwd
        self.manual_login = manual_login
        
    def getFriendsURL(self, my_url, store_in=None):
        """
        """
        file_name = store_in
        
        if os.path.isfile(file_name):
            with open(file_name, 'rb') as f:
                friends_url = pickle.load(f)
            print('We loaded {} uniq friends'.format(len(friends_url)))
            return friends_url

        friends_page = self.get_fb_page(my_url)
        parser = MyHTMLParser()
        parser.feed(friends_page)
        friends_url = set(parser.urls)

        print('We found {} friends, saving it...'.format(len(friends_url)))

        with open(file_name, 'wb') as f:
            pickle.dump(friends_url, f)
            
        return friends_url
    
    def login(self):
        """
        """
        user = self.user
        pwd = self.pwd
        driver = self.driver
        try:
            driver.get('http://www.facebook.com/')
            # authenticate to facebook account
            elem = driver.find_element_by_id("email")
            elem.send_keys(user)
            elem = driver.find_element_by_id("pass")
            elem.send_keys(pwd)
            elem.send_keys(Keys.RETURN)
            time.sleep(5)
            print("Successfully logged in Facebook!")
        except Exception as e:
            print(e)
    
    def download(self, login_required = False):
        """
        """
        SCROLL_PAUSE_TIME = self.SCROLL_PAUSE_TIME 
        manual_login = self.manual_login
        my_url = 'http://www.facebook.com/' + self.user + '/friends'
        
        if manual_login:
            driver = self.driver
            driver.get(my_url)#('http://www.facebook.com/')
            time.sleep(10)
            print("Have you logged in to the facebook account? (Y/N)")
            response = input()
            while response != 'Y':
                response = input()
        else: 
            self.login()
        
        my_url = 'http://www.facebook.com/' + self.user + '/friends'
        friends_url = self.getFriendsURL(my_url, store_in='data/friends_url.pickle')
        friends_network = self.getFriendsNetwork(friends_url, store_in='data/friends_network.pickle')
        
        
        return friends_url
    
    
    
    def get_fb_page(self, url):
        time.sleep(10)
        driver = self.driver
        SCROLL_PAUSE_TIME = self.SCROLL_PAUSE_TIME 
        driver.get(url)
        # Get scroll height
        last_height = driver.execute_script("return document.body.scrollHeight")

        while True:
            # Scroll down to bottom
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            # Wait to load page
            time.sleep(SCROLL_PAUSE_TIME)

            # Calculate new scroll height and compare with last scroll height
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
        html_source = driver.page_source
        return html_source
    
    def find_friend_from_url(self, url):
        if re.search('com\/profile.php\?id=\d+\&', url) is not None:
            m = re.search('com\/profile.php\?id=(\d+)\&', url)
            friend = m.group(1)
        else:
            m = re.search('com\/(.*)\?', url)
            friend = m.group(1)
        return friend
    
    
    def getFriendsNetwork(self, friends_url, store_in = None):
        # need to pause every 100 friends
        username = self.user
        friend_network = {}
        file_path = store_in

        if os.path.isfile(file_path):
            with open(file_path, 'rb') as f:
                friend_network = pickle.load(f)
            print('Loaded existing graph, found {} keys'.format(len(friend_network.keys())))

        count = 1
        for url in tqdm(friends_url):
            
            if count == 5:
                print(count,url)
                break
            
            if count % 100 == 0:
                print ("Too many queries, pause for a while...")
                time.sleep(1800)

            friend_username = self.find_friend_from_url(url)
            if (friend_username in friend_network.keys()) and (len(friend_network[friend_username]) >1):
                continue

            friend_network[friend_username] = [username]
            
            mutual_url = 'https://www.facebook.com/{}/friends_mutual'.format(friend_username)
            mutual_page = self.get_fb_page(mutual_url)
            parser = MyHTMLParser()
            parser.urls = []
            parser.feed(mutual_page)
            mutual_friends_urls = set(parser.urls)
            print('Found {} urls'.format(len(mutual_friends_urls)))
            
            for mutual_url in mutual_friends_urls:
                mutual_friend = self.find_friend_from_url(mutual_url)
                friend_network[friend_username].append(mutual_friend)

            with open(file_path, 'wb') as f:
                pickle.dump(friend_network, f)

            time.sleep(5)
            count += 1
```

```python
class FBNetworkAnalysis:
    def __init__(self):
        pass
    
    

    def convert(self, graph = None, to = 'gml'):
        """
        """
        My_ID = 'bilalansari.000'
        NETWORK_FILENAME = 'data/friends_network.pickle'

        with open(NETWORK_FILENAME, 'rb') as f:
            friends_network = pickle.load(f)
        
        friends_network = self.network_of(friends_network, minimum_degree = 4)
        
        edges = []
        nodes = [My_ID]

        edges = [(friend, matual_friend) 
                  for friend, matual_friends in friends_network.items() for matual_friend in matual_friends]
        
        print(edges,"\n\n",nodes)
        G = nx.Graph()
        G.add_nodes_from([My_ID])
        G.add_nodes_from(friends_network.keys())

        G.add_edges_from(edges)
        print('Added {} edges'.format(len(edges) ))
        
        write = {'gml' : nx.write_gml(G, 'data/friends_network.gml'),
                 'graphml': nx.write_graphml(G, 'data/friends_network.graphml'),}
        write[to]
        return 

    def network_of(self, friends_network, minimum_degree = 1):
        central_friends = {}
        for friend, matual_friends in friends_network.items():
            actual_number_matural_friends = len(np.intersect1d(list(friends_network.keys()), matual_friends))
            if actual_number_matural_friends >= minimum_degree:
                actual_matual_friends = np.intersect1d(list(friends_network.keys()), matual_friends)
                central_friends[friend] = actual_matual_friends
        return central_friends
        #print('Firtered out {} items'.format(len(friend_graph.keys()) - len(central_friends.keys())))
```

```python
import re
def find_friend_from_url(url):
    """Fetch the friend's facebook id from url"""
    if re.search('com\/profile.php\?id=\d+\&', url) is not None:
        m = re.search('com\/profile.php\?id=(\d+)\&', url)
        friend_id = m.group(1)
    else:
        m = re.search('com\/(.*)\?', url)
        friend_id = m.group(1)
    return friend_id
```

```python
import pandas as pd
import numpy as np
def get_basic_info(data):
    info = {'Birthday':np.NaN, 'Gender':pd.NA, 'Languages':pd.NA, 'Religious views':pd.NA, 'Political views':pd.NA}
    for info_type in info.keys():
        try:
            matches = list(re.finditer(info_type+'\n', data))
            info_matches = list(re.finditer("\n",data[matches[0].span()[1]:]))
            info[info_type] = data[matches[0].span()[1]:matches[0].span()[1]+info_matches[0].span()[1]-1]
        except:
            pass
    try:
        info['Languages'] = "^".join([l.strip() for l in info['Languages'].split('')])
    except:
        pass
    return info
```

```python
import pickle
import re
def fun():
    with open("data/more_info.pickle", 'rb') as f:
        more_info = pickle.load(f)
        m = {}
        for friend in more_info.keys():
            res = get_basic_info(more_info[friend]['basic_info']+"\n")
            res['Followers'] =  more_info[friend]['followers']
            res['Name'] =  more_info[friend]['name']
            m[friend] = res
        df = pd.DataFrame(m).transpose()
        return df
```
